```python
import pandas as pd
import numpy as np

data = {'Name':['arifa','fida','sana','minna'],'chem':[50,40,30,20],'phy':[60,70,80,90],'maths':[80,90,50,75]}

df = pd.DataFrame (data)

df['bio'] = [40,35,20,30]

df['total'] = df['chem'] +df['phy'] +df['maths'] +df['bio']

print(df)


list1 = []

for i in df['total']:

  if i<100:
    list1.append(' avg')

  elif i<=200:
   list1.append('above avg')

  else:
    list1.append('exnt')

df['status'] = list1

print(df[['Name','status','chem']])

df['status'] = list1

del df['chem']

df = df.drop('phy',axis=1)   #want to specify axis

#print(df[['name',.status','chem']])

df.pop('maths')

print(df)

print(df.ndim)
print(df.shape)
print(df.size)
print(df.dtypes)

print(df.columns)

print(df['status'].unique())

print(df['status'].value_counts())

print(df['bio'].unique())

print(df)
```
